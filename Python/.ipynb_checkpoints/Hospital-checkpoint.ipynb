```python
# change the Date column to dattime format
df.Date = pd.to_datetime(df.Date, format="%d/%m/%Y")
```

```python
df.head(1)
```

```python
# dataframe where code == NaN
x = df[df.Code.isna() == True]
```

```python
# get the unique value of the Name for its corresponding code is NaN
uniqueName = pd.unique(x.Name)
uniqueName
```

```python
# group by Name
groupName = x.groupby(by='Name')
```

```python
def populateCodeAndRegion(groupName, uniqueName):
    for i in range(len(uniqueName)):
        # for holding the 1st three letters of the Name
        Str = ''
        Str = uniqueName[i][:3]
        provideName = groupName.get_group(uniqueName[i])
        for j in provideName.index:
            df.loc[j, ["Code"]] = Str
            df.loc[j, ["Region"]] = Str+'_'+str(i)
```

```python
populateCodeAndRegion(groupName, uniqueName)
```
